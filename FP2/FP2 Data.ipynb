```python
import pandas as pd
import numpy as np
```

```python
d = pd.read_csv("investments_VC.csv", encoding= 'unicode_escape')
d
```

```python
d = d.dropna(subset=['status'], axis = 0)
d
```

```python
d.columns
```

```python
d = d.drop(['permalink', 'homepage_url', 'category_list', 'founded_month','founded_quarter','founded_year'], axis = 1)
d
```

```python
d.columns
```

```python
d = d[d["country_code"] == "USA"]
```

```python
d
```

```python
d = d[d["status"] != "operating"]
d
```

```python
# d.to_csv("investments_VC_cleaned", encoding='utf-8', index=False)
```

```python
d = d.rename({' market ': 'market', ' funding_total_usd ': 'funding_total_usd'}, axis=1)
```

```python
def formatFunding(x):
    val = x.strip()
    val = val.replace(",", "")
    return val
    
```

```python
d["funding_total_usd"] = d["funding_total_usd"].apply(lambda x: formatFunding(x))
d
```

```python
d = d[d["funding_total_usd"] != "-"]
```

```python
d = d.astype({'funding_total_usd': 'int64'})
```

```python
d = d.drop(["country_code", "region", "city"], axis=1)
```

```python
d
```

```python
d.columns
```

```python
d = d.astype({'market': 'str', 'name': 'str', 'status': 'str', 'state_code': 'str'})
```

```python
d
```

```python
d[["market", "name", "status", "state_code"]] = d[["market", "name", "status", "state_code"]].apply(lambda x: x.str.strip())
```
