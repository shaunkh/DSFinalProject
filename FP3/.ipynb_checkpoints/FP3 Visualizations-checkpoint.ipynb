```python
import plotly.express as px
```

```python
col_names
```

```python
fig = px.scatter(df, x = "market", y = "funding_total_usd")
fig.show()
```

```python
marketscount = df.groupby("market").count()
marketscount
```

```python
marketssorted = marketscount.sort_values(by="name", ascending=False)
marketssorted.head(10)
```

```python
fig = px.bar(marketssorted.head(10), y = "name",
            labels = {
                "name": "count"
            },
            title = "Top 10 Markets with Most Startups")
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
fig.show()
```

```python
topten = list(marketssorted.index.values)[:10]
topten
```

```python
markets = df.groupby("market")
markets.first()
```

```python
toptenmarketsbyfunding = markets["funding_total_usd"].sum().sort_values(ascending = False).head(10)
toptenmarketsbyfunding
```

```python
fig = px.bar(toptenmarketsbyfunding,
            labels = {
                "value": "Total Funding",
                "market": "Market"
            },
            title = "Top 10 Markets by Funding Amount")
fig.show()
```

```python
fig = px.scatter(df, x = "market", y = "status")
fig.show()
```

```python
markets2 = df.groupby(["market", "status"]).sum()
markets2
```

```python
markets3 = markets2.reset_index()
```

```python
fig = px.bar(markets3, x = "market", y = "funding_total_usd", color = "status",
            labels = {
                "funding_total_usd": "Total Funding",
                "market": "Market"
            },
            title = "Top 20 Markets by Funding Amount")
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
fig.update_xaxes(title_standoff = 75)
fig.show()
```

```python
fig = px.scatter(markets3, x = "funding_rounds", y = "funding_total_usd", color = "status",
            labels = {
                "funding_total_usd": "Total Funding",
                "market": "Market"
            },
            title = "")
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
fig.update_xaxes(title_standoff = 75)
fig.show()
```

```python
bystate = df.groupby(["state_code", "status"]).sum()
bystate = bystate.reset_index()
```

```python
fig = px.bar(bystate, x = "state_code", y = "funding_total_usd", color = "status",
            labels = {
                "funding_total_usd": "Total Funding",
                "state_code": "State"
            },
            title = "")
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'})
fig.update_xaxes(title_standoff = 75)
fig.show()
```

```python
fig = px.scatter(df, x = "debt_financing", y = "funding_total_usd", color = "status",
            title = "Debt Financing to Total Funding")
fig.show()
```
